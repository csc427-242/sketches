```python
#
# implementation of a necklace
# demonstration for csc427-242 univ of miami
# 23 jan 2024
# burton rosenberg
#

class NecklaceString:
    
    color = set(['w','b'])

    def __init__(self,necklace):
        self.check(necklace)
        self.necklace = necklace
    
    def check(self,n):
        assert type(n)==type(""), 'parameter is not a string'
        for b in n: 
            if b not in NecklaceString.color: 
                assert False, f'color {b} is not in {NecklaceString.color}'
        
    def rotate(self,i):
        i = i%len(self.necklace)
        return NecklaceString(self.necklace[-i:]+self.necklace[:-i])
    
    def orbit(self):
        s = set()
        for i in range(len(self.necklace)):
            s.add(self.rotate(i).necklace)
        return s
    
class NecklaceArray:
    
    color = set(['w','b'])

    def __init__(self,necklace):
        self.check(necklace)
        self.necklace = necklace
    
    def check(self,n):
        assert type(n)==type([]), 'parameter is not a list'
        for b in n: 
            if b not in NecklaceString.color: 
                assert False, f'color {b} is not in {NecklaceString.color}'
        
    def rotate(self,i):
        i = i%len(self.necklace)
        return NecklaceArray(self.necklace[-i:]+self.necklace[:-i])
    
    def orbit(self):
        s = []
        for i in range(len(self.necklace)):
            n = self.rotate(i).necklace
            if n not in s:
                s.append(n)
        return s
```

```python
n = 6

def all_strings(n):
    s = []
    if n<=1:
        return ['w','b']
    else:
        r = all_strings(n-1)
        for t in r:
            s += ['b'+t]
            s += ['w'+t]
    return s

a_s = all_strings(n)
print('necklace:\torbit\n-------------------------------')
for s in a_s:
    n = NecklaceString(s)
    o = n.orbit()
    print(f'{n.necklace}:\t\t{o}')
```
