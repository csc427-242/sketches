```python
#
# implementation of a necklace
# demonstration for csc427-242 univ of miami
# 23 jan 2024
# burton rosenberg
#

class NecklaceString:
    
    color = set(['w','b'])

    def __init__(self,necklace):
        self.check(necklace)
        self.necklace = necklace
    
    def check(self,n):
        assert type(n)==type(""), 'parameter is not a string'
        for b in n: 
            if b not in NecklaceString.color: 
                assert False, f'color {b} is not in {NecklaceString.color}'
        
    def rotate(self,i):
        i = i%len(self.necklace)
        return NecklaceString(self.necklace[-i:]+self.necklace[:-i])
    
    def orbit(self):
        s = set()
        for i in range(len(self.necklace)):
            s.add(self.rotate(i).necklace)
        return s
 
    @staticmethod
    def all_strings(n):
        s = []
        if n<=1:
            return ['w','b']
        else:
            r = NecklaceString.all_strings(n-1)
            for t in r:
                s += ['b'+t]
                s += ['w'+t]
        return s

    # thanks to Andrew Siciliano helping on this methd
    @staticmethod
    def all_orbits(n):
        a_o = set()
        for s in NecklaceString.all_strings(n):
            n = NecklaceString(s)
            o = frozenset(n.orbit())
            if o not in a_o:
                a_o.add(o)
        return a_o
```

```python
n = 6

print('necklace:\torbit\n-------------------------------')

for s in NecklaceString.all_strings(n):
    nk = NecklaceString(s)
    o = nk.orbit()
    print(f'{nk.necklace}:\t\t{o}')
    
print()
print('size:\torbit\n-------------------------------')
for o in NecklaceString.all_orbits(n):
    print(f'{len(o)}:\t',end='')
    for c in o:
        print(f'{c} ',end='')
    print()
```
